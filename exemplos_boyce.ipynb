```python
from sympy import var, plot_implicit, Eq
```

```python
# exemplo 2, Boyce pg 34
var('x y')
fig = plot_implicit(Eq(y**2-2*y-x**3-2*x**2-2*x,3),
                    line_color='red',
                    show=False)

for c in range(-10,10,2):
    eq = Eq(y**2-2*y-x**3-2*x**2-2*x, c)
    fig.append(plot_implicit(eq,show=False)[0])
fig.show()
#print(fig)
#eq = Eq(y**2-2*y-x**3-2*x**2-2*x, 3)
#fig.extend(plot_implicit(eq,line_color = "red", show=False))
```

## Exemplo 3 (Boyce, p. 108)

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Exemplo 2 (Boyce, p. 34)
t = sym.symbols('t')
y = sym.Function('y')(t)
eq = y.diff(t,t)+5*y.diff(t)+6*y
sym.dsolve(eq, y)
```

```python
# o PVI
sol = sym.dsolve(eq, y, ics={y.subs(t, 0):2, y.diff(t).subs(t, 0): 3})
sol
```

```python
# sol.rhs  o lado direito da equao sol
f = sym.lambdify(t,sol.rhs)
```

```python
t = np.linspace(0,2)
plt.plot(t, f(t))
plt.grid()
plt.show()
```

```python
sols = lambda t,c1,c2: (c1-c2*np.exp(-t))*np.exp(-2*t)

for c1 in np.arange(-20,10,5):
    for c2 in np.arange(-20,20,5):
        plt.plot(t, sols(t,c1,c2),'b-',lw=0.5)
plt.grid()
plt.show()
```

## Exemplo 1 (Boyce, p. 122)

```python
t = sym.symbols('t')
y = sym.Function('y')(t)
eq = y.diff(t,t)+y.diff(t)+9.25*y
sym.dsolve(eq, y)
```

```python
sol = sym.dsolve(eq, y, ics={y.subs(t, 0):2, y.diff(t).subs(t, 0): 8})
sol
```

```python
# sol.rhs  o lado direito da equao sol
# lambdify converte em uma funo lambda
f = sym.lambdify(t,sol.rhs)
```

```python
t = np.linspace(0,10,100)
plt.plot(t, f(t))
plt.grid()
plt.show()
```

Variando $c_1$ e $c_2$

```python
sols = lambda t,c1,c2: (c1*np.sin(3.0*t) + c2*np.cos(3.0*t))*np.exp(-0.5*t)
t = np.linspace(0,4,100)

for c1 in range(0,100,20):
    for c2 in range(0,100,20):
        plt.plot(t, sols(t, c1,c2), 'b-', lw=0.3)
        plt.plot(t, sols(t, c1,c2), 'r:', lw=0.3)
plt.grid()
plt.show()
```

## Exemplo 2 (Boyce, p. 124)

```python
t = sym.symbols('t')
y = sym.Function('y')(t)
eq = 16*y.diff(t,t)-8*y.diff(t)+145*y
sol_g = sym.dsolve(eq, y)
sol_g
```

```python
sol_p = sym.dsolve(eq, y, ics={y.subs(t, 0):-2, y.diff(t).subs(t, 0): 1})
sol_p
```

```python
f = sym.lambdify(t,sol_p.rhs)
t = np.linspace(0,10,100)
plt.plot(t, f(t))
plt.grid()
plt.show()
```
