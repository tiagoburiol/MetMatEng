# 4. Solução em série para EDOs de 2ª ordem

**Exemplo 1:** Vamos encontrar uma solução em série de potência para $y''+y=0$

Buscamos uma solução na forma $$y = \sum_{n=0}^\infty a_nx^n$$

derivando duas vezes obtemos

$$y' = \sum_{n=1}^\infty na_nx^{n-1} \,\,\,\,\text{e}\,\,\,\, y'' = \sum_{n=2}^\infty n(n-1)a_nx^{n-2}  $$

substituindo na equação $y''+y=0$ obtemos

$$\sum_{n=2}^\infty n(n-1)a_nx^{n-2} + \sum_{n=0}^\infty a_nx^n =0$$ 

fazendo $n$ começar em $0$ nas duas somas ($n=n+2$ nos termos da primeira soma) obtemos

$$\sum_{n=0}^\infty (n+2)(n+1)a_{n+2}x^n + \sum_{n=0}^\infty a_nx^n=0$$ 

ou

$$\sum_{n=0}^\infty \left[(n+2)(n+1)a_{n+2} + a_n \right]x^n=0$$ 

igualando os coeficientes das potências de $x$ de mesmo grau chegamos a

$$(n+2)(n+1)a_{n+2} + a_n = 0$$

ou 

$$a_{n+2} = -\frac{a_n}{(n+2)(n+1)}$$

fazendo $n=0,1,2,3,4,...$ obtemos

$$a_2=-\frac{a_0}{2!}, a_3=-\frac{a_1}{3!}, a_4=\frac{a_0}{4!}, a_5=\frac{a_1}{5!}, a_6=-\frac{a_0}{6!}, a_7=-\frac{a_1}{7!},...$$

Assim, a solução pode ser escrita como

$$ y(x) = a_0+a_1x-\frac{a_0}{2!}x^2-\frac{a_1}{3!}x^3+\frac{a_0}{4!}x^4+ a_5=\frac{a_1}{5!}x^5-\frac{a_0}{6!}x^6-\frac{a_1}{7!}x^7+...$$

ou

$$ y(x)= a_0\sum_{n=0}^\infty \frac{(-1)^n}{2n!}x^{2n} + a_1\sum_{n=0}^\infty \frac{(-1)^n}{(2n+1)!}x^{2n+1}$$

O resultado obtido é equivalente ao resultado já conhecido

$$y(x) = a_0 cos(x) + a_1 sen(x)$$

Os códigos abaixo geram os gráficos das séries para $cos(x)$ e $sen(x)$ com diferentes números de termor e no fim algumas das soluções para alguns valores de $a_0$ e $a_1$

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
par = 1
impar = 0
N = 10
x = sym.symbols('x') 
xi = np.linspace(0,10)

plt.figure(figsize = (10, 4))
plt.plot(xi, np.cos(xi), 'k--')
plt.ylim(-2,2)
plt.xlim(0,10)
plt.grid()

for n in range(1,N):
    par += (-1)**n*x**(2*n)/sym.factorial(2*n)
    s = sym.lambdify(x,par)
    plt.plot(xi,s(xi),lw=0.7)
par
```

**Exemplo 2:** Vamos encontrar uma solução em série de potência para $y''+xy=0$

Novamente buscamos uma solução na forma $$y = \sum_{n=0}^\infty a_nx^n$$

derivando duas vezes e substituino na equação ontemos

$$\sum_{n=2}^\infty n(n-1)a_nx^{n-2} = \sum_{n=0}^\infty a_nx^{n+1} =0$$ 

fazendo $n$ começar em $1$ nas duas somas ($n=n-1$ nos termos da soma a direita) obtemos

$$2a_2+\sum_{n=1}^\infty (n+2)(n+1)a_{n+2}x^n = \sum_{n=1}^\infty a_{n11}x^n=0$$ 

igualando os coeficientes das potências de $x$ de mesmo grau chegamos a

$$(n+2)(n+1)a_{n+2} = a_{n-1}$$

ou 

$$a_{n+2} = -\frac{a_n}{(n+2)(n+1)}$$

para $n=1,2,3,4,...$. Calculando os coeficientes em função de $a_0$ e $a_1$, a solução geral fica na forma
